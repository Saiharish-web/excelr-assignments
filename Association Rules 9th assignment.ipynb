```python
dataset
transaction=[]
for i in  range(0,2000):
    transaction.append([str(dataset.values[i,u])for u in range(1,10)])
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_books = apriori(dataset, min_support=0.05, max_len=3,use_colnames = True)
frequent_books
```

```python
frequent_books.sort_values('support',ascending = False,inplace=True)
frequent_books.sort_values
```

```python
rules = association_rules(frequent_books, metric="lift", min_threshold=1)
```

```python
rules
```

```python
rules.sort_values('lift',ascending=False,inplace=True)
```

```python
import pandas as pd 
import numpy as np 
```

```python
movies = pd.read_csv("my_movies.csv",sep=',')
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies = movies.replace(np.nan, '', regex=True)
movies
```

```python
movies
transactions = []
for i in range(0, 10):
    transactions.append([str(movies.values[i,u]) for u in range(1,5)])
```

```python
movies_series  = pd.DataFrame(pd.Series(transactions))
movies_series
movies_series.columns = ["business"]
movies_series
```

```python
X = movies_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_movies = apriori(X, min_support=0.1, max_len=4,use_colnames = True)
frequent_movies 
```

```python
frequent_movies.sort_values('support',ascending = False,inplace=True)
frequent_movies.sort_values
```
